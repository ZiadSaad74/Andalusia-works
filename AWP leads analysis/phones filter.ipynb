```python
import pandas as pd 
import numpy as np
import warnings
```

```python
df = pd.read_excel('C:\\Users\\ziad.saad\\Desktop\\All Phone Calls.xlsx')

df['Created On'] = pd.to_datetime(df['Created On'])
df = df[df['Created On'] >= '10/01/2024  00:00:00']
df = df[df['Business Unit (Owning User) (User)'] =='AWP']

#df.head(2)
```

```python
def check(df):
    data_info = []
    cols = df.columns
    for col in cols:
        name = col
        dtypes = df[col].dtype
        valuesCounts = df[col].value_counts().sum()
        nulls = df[col].isnull().sum()
        uniques = df[col].nunique()
        column = ['Name', 'Data types','Nulls count','Value counts','Uniques count']
        data_info.append([name, dtypes,nulls, valuesCounts, uniques])
    data = pd.DataFrame(data_info)
    data.columns = column
    return data
```

```python
check(df)
```

```python
def remove_cols(df,cols):
    return df.drop(cols, axis=1, inplace=True)

col = ['(Do Not Modify) Phone Call','(Do Not Modify) Row Checksum','(Do Not Modify) Modified On','Call To','Description', 'Source Campaign',
       'Business Unit (Owning User) (User)','Created By','Due', 'Call From','Country (Regarding) (Contact)','Department (Owning User) (User)',
       'Site (Owning User) (User)']

remove_cols(df,col)
```

```python
def rename(df, olds,news):
    dict_names = {}
    for i in range(len(olds)):
        dict_names[olds[i]] = news[i]
    return df.rename(columns=dict_names, inplace=True )

olds = ["Regarding",'Subject','Mobile Last 5 digits (Regarding) (Lead)','Created On']
news = ["Name",'Topic','Mobile Phone','first call']
rename(df,olds,news)
```

```python
cols = ['Topic', 'Name', 'Mobile Phone', 'Owner','Phone Call Result', 'Not Intersted Reason','Follow Up Action', 'Follow Up Date',
        'Next Step','first call', 'Priority', 'Activity Status', 'Direction']

df = df[cols]
df.head(1)
```

```python
c = 0

for i in df['Phone Call Result']:
    if pd.isna(df['Not Intersted Reason'].loc[c]):
        
        if i == 'Wrong No':
            df.loc[c,'Not Intersted Reason'] = 'Wrong Number'

        elif i in ['Call Droped', 'Switched Off','Left a Message','No Answer','Busy Call Back Later']:
            df.loc[c,'Not Intersted Reason'] = 'Not reached call'

        elif i == 'Interested':
            df.loc[c,'Not Intersted Reason'] = 'Interested'

        elif i == 'Booked':
            df.loc[c,'Not Intersted Reason'] = 'Booked'            

        elif i in ['Complain','Unsure']:
            df.loc[c,'Not Intersted Reason'] = 'Not convinced with Business Idea'

        elif i in ['Interested Follow Up Required','Interested but not available'] :
            df.loc[c,'Not Intersted Reason'] = 'Need Next Step'

        elif i == 'Low Budget':
            df.loc[c,'Not Intersted Reason'] = 'High Price'

        elif i == 'Cancel EOI':
            df.loc[c,'Not Intersted Reason'] = 'Cancelled EOI'

        elif i == 'Not Interested':
            df.loc[c,'Not Intersted Reason'] = 'Unknown'

    c+=1
```

```python
c = 0

for i in df['Phone Call Result']:
    if pd.isna(df['Follow Up Action'][c]):    
        if i == 'Not Interested':
            df.loc[c,'Follow Up Action'] = 'No Need'
    c+=1
```

```python
c = 0

for i in df['Follow Up Action']:
    if i == 'No Need':
        df.loc[c,'Next Step'] = i
    c+=1
```
